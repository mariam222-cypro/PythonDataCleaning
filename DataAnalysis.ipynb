### Step 0 

1.   Load the energy data from the file `Energy Indicators.xls` , list of indicators of [energy supply and renewable electricity production](Energy%20Indicators.xls) from the [United Nations](http://unstats.un.org/unsd/environment/excel_file_tables/2013/Energy%20Indicators.xls) for the year 2013, and should be put into a DataFrame with the variable name of **energy**.

2.   Exclude the footer and header information from the datafile.
3.   Get rid of the first two columns 
4.   Change the column labels so that the columns are:
`['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']`

5. Convert `Energy Supply` to gigajoules (there are 1,000,000 gigajoules in a petajoule). 
6. For all countries which have missing data (e.g. data with "...") make sure this is reflected as `np.NaN` values.

7. Rename the following list of countries (for use in later questions):
```"Republic of Korea": "South Korea",
"United States of America": "United States",
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
"China, Hong Kong Special Administrative Region": "Hong Kong"```

8. There are also several countries with numbers and/or parenthesis in their name. Be sure to remove these, 

  e.g. 
```"Bolivia (Plurinational State of)" should be "Bolivia", 
"Switzerland17" should be "Switzerland"```.

````python
#Import Lib
import pandas as pd
import requests
import numpy as np
````

````python
energy = pd.read_excel('Energy Indicators.xls',header = None,skiprows=18,skipfooter= 38,usecols=[2,3,4,5],
                       names=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable'])

def remove(x):
  for i in range(len(x)):
    if x[i].isdigit():
      return x[:i]
    elif (x[i]=='('):
      return x[:i-1]
  return x

energy['Country'] = energy['Country'].apply(remove)    

energy['Energy Supply'] = energy['Energy Supply'].apply(lambda x: x*1000000)

energy['Country'] = energy['Country'].replace({'Republic of Korea' : 'South Korea',
                                               'United States of America' : 'United States',
                                               'United Kingdom of Great Britain and Northern Ireland':'United Kingdom',
                                               'China, Hong Kong Special Administrative Region':'Hong Kong',
                                               'Bolivia (Plurinational State of)': 'Bolivia'})
energy = energy.replace('...','np.NaN')
````

### Step 9



*   Create a new column with **a 1** if the country's % Renewable value is **at or above the median** for all countries in the top 15, **and a 0** if the country's % Renewable value **is below the median**.



*This step should return a series named `HighRenew` whose index is the country name sorted in ascending order of rank.*

````python
def get_Median(x):
  median_value=dfFinal["% Renewable"].median()
  if x >= median_value : 
    return 1
  else :
    return 0

dfFinal["HighRenew"]= dfFinal["% Renewable"].apply(get_Median).sort_values(ascending=True)

dfFinal["HighRenew"]
````
